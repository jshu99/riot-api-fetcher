```python
import pandas as pd

# Load your data
df = pd.read_csv('player_stats_new.csv')

# Find duplicate player entries in same match
duplicates = df[df.duplicated(subset=['matchId', 'riotIdGameName'], keep=False)]

# Display the duplicate entries
print("Duplicate player entries in same match:")
print(f"Total duplicate entries: {len(duplicates)}")
print("\n" + "="*80)

# Show duplicates sorted by matchId and player name for easy inspection
duplicates_sorted = duplicates.sort_values(['matchId', 'riotIdGameName'])

# Display key columns to see the duplicates clearly
print(duplicates_sorted[['matchId', 'riotIdGameName', 'championName', 'kills', 'deaths', 'assists', 'teamPosition']].to_string(index=False))

# Show which specific matches have duplicates
print("\n" + "="*80)
print("Matches with duplicate player entries:")
match_counts = duplicates.groupby('matchId')['riotIdGameName'].nunique()
print(match_counts)

# Show the specific players that are duplicated
print("\n" + "="*80)
print("Players with duplicate entries:")
player_counts = duplicates.groupby(['matchId', 'riotIdGameName']).size()
print(player_counts)
```

```python
import pandas as pd
import numpy as np

# Load your data
df = pd.read_csv('player_stats_new.csv')

# Method 1: Check for null/NaN values
print("=== NULL/NaN VALUES ===")
print(f"Total null values: {df.isnull().sum().sum()}")
print(f"Rows with any null values: {df.isnull().any(axis=1).sum()}")

# Show null counts by column
print("\nNull values by column:")
null_counts = df.isnull().sum()
print(null_counts[null_counts > 0])  # Only show columns with nulls

# Method 2: Check for empty strings
print("\n=== EMPTY STRINGS ===")
empty_strings = (df == '').sum().sum()
print(f"Total empty strings: {empty_strings}")

# Check empty strings by column
empty_string_counts = (df == '').sum()
print("\nEmpty strings by column:")
print(empty_string_counts[empty_string_counts > 0])

# Method 3: Check for whitespace-only strings
print("\n=== WHITESPACE-ONLY STRINGS ===")
whitespace_only = df.apply(lambda x: x.astype(str).str.strip() == '').sum().sum()
print(f"Total whitespace-only strings: {whitespace_only}")

# Method 4: Show rows with missing data
print("\n=== ROWS WITH MISSING DATA ===")
rows_with_nulls = df[df.isnull().any(axis=1)]
print(f"Sample of rows with null values:")
print(rows_with_nulls.head())

# Method 5: Detailed analysis by column
print("\n=== DETAILED ANALYSIS BY COLUMN ===")
for column in df.columns:
    null_count = df[column].isnull().sum()
    empty_count = (df[column] == '').sum()
    total_missing = null_count + empty_count
    
    if total_missing > 0:
        print(f"{column}:")
        print(f"  - Null values: {null_count}")
        print(f"  - Empty strings: {empty_count}")
        print(f"  - Total missing: {total_missing}")
        print(f"  - Percentage missing: {(total_missing/len(df)*100):.2f}%")

# Method 6: Check specific columns that should never be empty
critical_columns = ['matchId', 'riotIdGameName', 'championName', 'kills', 'deaths', 'assists']
print("\n=== CRITICAL COLUMNS CHECK ===")
for col in critical_columns:
    if col in df.columns:
        missing = df[col].isnull().sum() + (df[col] == '').sum()
        print(f"{col}: {missing} missing values")

# Method 7: Show complete rows vs incomplete rows
complete_rows = df.dropna().shape[0]
incomplete_rows = len(df) - complete_rows
print(f"\n=== COMPLETENESS SUMMARY ===")
print(f"Complete rows: {complete_rows}")
print(f"Incomplete rows: {incomplete_rows}")
print(f"Completeness rate: {(complete_rows/len(df)*100):.2f}%")
```

```python
import pandas as pd

# Load your data
df = pd.read_csv('player_stats_new.csv')

# Check for null teamPosition values
print("=== TEAM POSITION NULL CHECK ===")
null_team_position = df[df['teamPosition'].isnull()]
print(f"Rows with null teamPosition: {len(null_team_position)}")

if len(null_team_position) > 0:
    print("\nRows with null teamPosition:")
    print(null_team_position[['matchId', 'riotIdGameName', 'championName', 'teamPosition', 'kills', 'deaths', 'assists']].to_string(index=False))
    
    # Show unique matches that have null teamPosition
    matches_with_null_position = null_team_position['matchId'].unique()
    print(f"\nUnique matches with null teamPosition: {len(matches_with_null_position)}")
    print("Match IDs:", matches_with_null_position.tolist())
    
    # Check if these are all from the same match
    if len(matches_with_null_position) == 1:
        print(f"\nAll null teamPosition values are from match: {matches_with_null_position[0]}")
    
    # Show the full data for these rows
    print("\nFull data for rows with null teamPosition:")
    print(null_team_position.to_string(index=False))

else:
    print("No null teamPosition values found!")

# Additional analysis
print("\n=== TEAM POSITION ANALYSIS ===")
print("Team position value counts:")
print(df['teamPosition'].value_counts(dropna=False))

print("\nPercentage of null teamPosition:")
null_percentage = (df['teamPosition'].isnull().sum() / len(df)) * 100
print(f"{null_percentage:.2f}%")
```

```python
import pandas as pd

# Load your data
df = pd.read_csv('player_stats.csv')

# Check for matches with more than 10 players
print("=== MATCHES WITH MORE THAN 10 PLAYERS ===")

# Count players per match
players_per_match = df.groupby('matchId').size()
matches_with_extra_players = players_per_match[players_per_match > 10]

print(f"Matches with more than 10 players: {len(matches_with_extra_players)}")

if len(matches_with_extra_players) > 0:
    print("\nMatches with extra players:")
    for match_id, player_count in matches_with_extra_players.items():
        print(f"Match {match_id}: {player_count} players")
        
        # Show the players in this match
        match_data = df[df['matchId'] == match_id]
        print(f"Players in {match_id}:")
        print(match_data[['riotIdGameName', 'championName', 'teamPosition', 'kills', 'deaths', 'assists']].to_string(index=False))
        print("-" * 80)

else:
    print("No matches found with more than 10 players!")

# Additional analysis
print("\n=== PLAYER COUNT ANALYSIS ===")
print("Player count distribution:")
player_count_dist = players_per_match.value_counts().sort_index()
print(player_count_dist)

print(f"\nTotal matches: {len(players_per_match)}")
print(f"Matches with exactly 10 players: {len(players_per_match[players_per_match == 10])}")
print(f"Matches with fewer than 10 players: {len(players_per_match[players_per_match < 10])}")
print(f"Matches with more than 10 players: {len(players_per_match[players_per_match > 10])}")

# Check for matches with fewer than 10 players too
matches_with_fewer_players = players_per_match[players_per_match < 10]
if len(matches_with_fewer_players) > 0:
    print(f"\nMatches with fewer than 10 players: {len(matches_with_fewer_players)}")
    print("These might be incomplete data or early surrender games:")
    for match_id, player_count in matches_with_fewer_players.head(5).items():
        print(f"Match {match_id}: {player_count} players")

# Show summary statistics
print("\n=== SUMMARY STATISTICS ===")
print(f"Average players per match: {players_per_match.mean():.2f}")
print(f"Median players per match: {players_per_match.median()}")
print(f"Min players per match: {players_per_match.min()}")
print(f"Max players per match: {players_per_match.max()}")
```

```python
import pandas as pd

# Load your data
df = pd.read_csv('old/player_stats.csv')

# Check Blitzcrank's team positions
print("=== BLITZCRANK TEAM POSITIONS ===")

# Filter for Blitzcrank games
blitzcrank_games = df[df['championName'] == 'Blitzcrank']

if len(blitzcrank_games) > 0:
    print(f"Total Blitzcrank games: {len(blitzcrank_games)}")
    
    # Show team position distribution
    print("\nTeam position distribution:")
    position_counts = blitzcrank_games['teamPosition'].value_counts()
    print(position_counts)
    
    # Show percentage breakdown
    print("\nTeam position percentage:")
    position_percentages = blitzcrank_games['teamPosition'].value_counts(normalize=True) * 100
    for position, percentage in position_percentages.items():
        print(f"{position}: {percentage:.1f}%")
    
    # Show sample games for each position
    print("\nSample games by position:")
    for position in blitzcrank_games['teamPosition'].unique():
        if pd.notna(position):  # Check for non-null values
            sample_games = blitzcrank_games[blitzcrank_games['teamPosition'] == position].head(3)
            print(f"\n{position} position (sample):")
            print(sample_games[['matchId', 'riotIdGameName', 'teamPosition', 'kills', 'deaths', 'assists', 'win']].to_string(index=False))
    
    # Check for any null team positions
    null_positions = blitzcrank_games[blitzcrank_games['teamPosition'].isnull()]
    if len(null_positions) > 0:
        print(f"\nBlitzcrank games with null team position: {len(null_positions)}")
        print(null_positions[['matchId', 'riotIdGameName', 'teamPosition', 'kills', 'deaths', 'assists']].to_string(index=False))

else:
    print("No Blitzcrank games found in the dataset!")

# Additional analysis - check if there are any unusual positions
print("\n=== UNUSUAL POSITIONS CHECK ===")
unusual_positions = blitzcrank_games[~blitzcrank_games['teamPosition'].isin(['UTILITY', 'BOTTOM', 'MIDDLE', 'TOP', 'JUNGLE'])]
if len(unusual_positions) > 0:
    print("Blitzcrank games with unusual positions:")
    print(unusual_positions[['matchId', 'riotIdGameName', 'teamPosition', 'kills', 'deaths', 'assists']].to_string(index=False))
else:
    print("All Blitzcrank positions are standard (UTILITY, BOTTOM, MIDDLE, TOP, JUNGLE)")
```
